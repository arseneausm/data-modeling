```python
import pandas as pd
import numpy as np
from numpy.linalg import inv

import matplotlib.pyplot as plt
```

```python
dat = pd.read_csv('data/01_standard_practice.csv')

dat
```

```python
def linear(x,m,b):
    return m*x+b

def fit_standard(x, y, sig_x = None, sig_y = None):
    if sig_x is None:
        sig_x = np.zeros(len(x))
    if sig_y is None:
        sig_y = np.zeros(len(y))
        
    A = np.array([[1, x[n]] for n in range(len(x))])
    C = np.diagflat(sig_y**2)
        
    cov = inv(A.T @ inv(C) @ A)
        
    X = cov @ ( A.T @ inv(C) @ y )
    
    return X, cov

x = np.array(dat['x'][4:])
sig_x = np.array(dat['sig_x'][4:])
y = np.array(dat['y'][4:])
sig_y = np.array(dat['sig_y'][4:])

X, cov = fit_standard(x, y, sig_x, sig_y)

print('==== fit report ====\n  m = %2.2f +/- %2.2f\n  b = %2.2f +/- %2.2f' % (X[1], np.sqrt(cov[1][1]), X[0], np.sqrt(cov[0][0])))

plt.figure(figsize=(8,5))

plt.errorbar(x, y, yerr = sig_y, color = 'black', fmt = 'o')
plt.plot(x, linear(x, X[1], X[0]))

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
x = np.array(dat['x'])
sig_x = np.array(dat['sig_x'])
y = np.array(dat['y'])
sig_y = np.array(dat['sig_y'])

X, cov = fit_standard(x, y, sig_x, sig_y)

print('==== fit report ====\n  m = %2.2f +/- %2.2f\n  b = %2.2f +/- %2.2f' % (X[1], np.sqrt(cov[1][1]), X[0], np.sqrt(cov[0][0])))

plt.figure(figsize=(8,5))

plt.errorbar(x, y, xerr = sig_x, yerr = sig_y, color = 'black', fmt = 'o')

xmin, xmax = plt.xlim()
dom = np.linspace(xmin, xmax, 50)
plt.plot(dom, linear(dom, X[1], X[0]))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
def quadratic(x, q, m, b):
    return q * x**2 + m * x + b

def fit_quadratic(x, y, sig_x = None, sig_y = None):
    if sig_x is None:
        sig_x = np.zeros(len(x))
    if sig_y is None:
        sig_y = np.zeros(len(y))
        
    A = np.array([[1, x[n], x[n]**2] for n in range(len(x))])
    C = np.diagflat(sig_y**2)
        
    cov = inv(A.T @ inv(C) @ A)
        
    X = cov @ ( A.T @ inv(C) @ y )
    
    return X, cov

x = np.array(dat['x'][4:])
sig_x = np.array(dat['sig_x'][4:])
y = np.array(dat['y'][4:])
sig_y = np.array(dat['sig_y'][4:])

X, cov = fit_quadratic(x, y, sig_x, sig_y)

print(X)

print('==== fit report ====\n  q = %2.4f +/- %2.4f\n  m = %2.2f +/- %2.2f\n  b = %2.2f +/- %2.2f' % (X[2], np.sqrt(cov[2][2]), X[1], np.sqrt(cov[1][1]), X[0], np.sqrt(cov[0][0])))

plt.figure(figsize=(8,5))

plt.errorbar(x, y, yerr = sig_y, color = 'black', fmt = 'o')

xmin, xmax = plt.xlim()
dom = np.linspace(xmin, xmax, 50)
plt.plot(dom, quadratic(dom, X[2], X[1], X[0]))

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
